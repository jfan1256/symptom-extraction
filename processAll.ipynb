## Move files that end with _ans.csv to new folder

```python
def moveAnswer(root):
    for path, subdirs, files in os.walk(root):
        for name in files:
            print(name)
            nameString = str(name)
            if nameString.endswith('_ans.csv'):
                shutil.move('./annotations/'+ name, './annotationsAns')
```

## Defining function to create data of single file

```python
def createData(recordFile, symptomFile):

    #Opens record and splits each line
    with open (recordFile, "r") as myfile:
        data = myfile.read().splitlines()
    #Opens raw record
    with open (recordFile, "r") as myfile:
        data1 = myfile.read()
    
    #Creates library of symptoms from _ref.csv
    recordRef = pd.read_csv(symptomFile)
    end = 0
    library = []
    for start in recordRef['START_CHAR']:
        endPos = recordRef['END_CHAR'][end]
        library.append(data1[start:endPos])
        end = end + 1
        
    #Creates list ['word', 'start index', 'end index]
    index = []
    prevIndex = 0
    x = 0
    for line in data:
        split = line.split()
        if split==[]:
            split = ['']
        for i in split:
            if x == 0:
                index.append([i, str(data1.index(i)), str(data1.index(i)+len(i))])
                prevIndex = data1.index(i)+len(i)
                x = 1
            elif i == '' and x==1:
                prevIndex = prevIndex+1
            elif i != [] and x==1:
                index.append([i, str(prevIndex+1), str(prevIndex+1+len(i))])
                prevIndex = prevIndex+1+len(i)
    
    #Creates list ['beginning word of each line']
    sentenceStart = []
    for line in data:
        split = line.split()
        if split == []:
            continue
        else:
            sentenceStart.append(split[0])
    
    #Creates dataframe
    df = pd.DataFrame(columns = ['Sentence', 'Word', 'Tag'])
    number = 1
    index0 = 0
    check = 0 
    for word in index:
        sentence = 'Sentence: ' + str(number)
        insert = 0
        check = 0
        index1 = 0
        for start in recordRef['START_CHAR']:
            end = recordRef['END_CHAR'][index1]
            index1 = index1 + 1
            if int(word[1]) == start or int(word[2]) == end or (start<int(word[1])<end):
                if sentenceStart[index0] == word[0]:
                    df.loc[len(df.index)] = [sentence, word[0], '1']
                    insert = 1
                    index0=index0+1
                    number = number + 1
                    check = 1
                    break
                else: 
                    df.loc[len(df.index)] = ['', word[0], '1']
                    insert = 1
                    check = 1
                    break
        #if check == 0:
        #    for symptom in library:
        #        if word[0] == symptom:
        #            if sentenceStart[index0] == word[0]:
        #                df.loc[len(df.index)] = [sentence, word[0], '1']
        #                insert = 1
        #                index0=index0+1
        #                number = number + 1
        #                break
        #            else: 
        #                df.loc[len(df.index)] = ['', word[0], '1']
        #                insert = 1
        #                break
        
        if insert == 0:
            if sentenceStart[index0] == word[0]:
                df.loc[len(df.index)] = [sentence, word[0], '0']
                index0=index0+1
                number = number + 1
            else: 
                df.loc[len(df.index)] = ['', word[0], '0']
    
    #Removes unnecessary characters/words
    counters = 0
    for words in df['Word']:
        if df['Sentence'][counters].__contains__('Sentence') == True:
            if words.__contains__('|') or words.__contains__(',') or words.__contains__('?') or words.__contains__('&') or words.__contains__('*') or words.__contains__('(') or words.__contains__(')') or words.__contains__('-') or words.__contains__('1') or words.__contains__('2') or words.__contains__('3') or words.__contains__('4') or words.__contains__('5') or words.__contains__('6') or words.__contains__('7') or words.__contains__('8') or words.__contains__('9') or words.__contains__('@') or words.__contains__('[') or words.__contains__(']') or words.__contains__('#') or words.__contains__(';') or len(words)<2:
                df['Sentence'][counters+1] = df['Sentence'][counters]
                df = df.drop(labels=counters, axis=0)
        else:
            if words.__contains__('|') or words.__contains__(',') or words.__contains__('?') or words.__contains__('&') or words.__contains__('*') or words.__contains__('(') or words.__contains__(')') or words.__contains__('-') or words.__contains__('1') or words.__contains__('2') or words.__contains__('3') or words.__contains__('4') or words.__contains__('5') or words.__contains__('6') or words.__contains__('7') or words.__contains__('8') or words.__contains__('9') or words.__contains__('@') or words.__contains__('[') or words.__contains__(']') or words.__contains__('#') or words.__contains__(';') or len(words)<2:
                df = df.drop(labels=counters, axis=0)
        counters = counters + 1
    return df
```

## Defining funciton for batch process to create dataset

```python
def scan_folder(recordFolder, symptomFolder):

    totalData = pd.DataFrame(columns = ['Sentence', 'Word', 'Tag'])
    count = 1
    for path, subdirs, files in os.walk(recordFolder):
        for nameRec in files:
            recordSRC = pathlib.Path(path, nameRec)
            symptomSRC = ''
            needBreak = 0
            for path1, subdirs1, files1 in os.walk(symptomFolder):
                for nameSym in files1:
                    if nameSym.startswith(nameRec):
                        symptomSRC = pathlib.Path(path1, nameSym)
                        needBreak = 1
                        break
                if needBreak == 1:
                    break

            try:
                res = createData(recordSRC, symptomSRC)
                totalData = totalData.append(res)
                print(recordSRC)
                print(symptomSRC)
                print(count)
                print('-----------------------------------------------')
                count = count+1
            except: 
                continue
    print('DONE')
    return totalData
    
        
                
```
